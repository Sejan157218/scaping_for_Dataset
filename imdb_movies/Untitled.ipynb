```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

movie_name = []
genre = []
certificate=[]
year = []
runtime = []
rating = []
metascore = []
budget=[]
gross = []
description = []
director = []
rating = []
writer_data = []
star_data = []

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36"}

url = 'https://www.imdb.com/title/tt17351924/?ref_=sr_t_1'

#request allow you to send HTTP request
response = requests.get(url,headers=headers)

soup = BeautifulSoup(response.content, 'html.parser')
```

```python
content_div = soup.find('div', class_="sc-e226b0e3-3 dwkouE")

print(content_div.findAll('li', class_='ipc-inline-list__item')[-1].text)

try:
    movie_name.append(content_div.find('span', class_='hero__primary-text').text)
except:
    movie_name.append(None)
try:
    year.append(content_div.find('a', class_='ipc-link ipc-link--baseAlt ipc-link--inherit-color').text)
except:
    year.append(None)

try:
    certificate.append(content_div.find('a', class_='ipc-link ipc-link--baseAlt ipc-link--inherit-color').text)
except:
    certificate.append(None)
    
try:
    runtime.append(content_div.find('li', class_='ipc-inline-list__item').text)
except:
    runtime.append(None)
    
try:
    rating.append(content_div.find('span', class_='sc-bde20123-1 cMEQkK').text)
except:
    rating.append(None)
    
try:
    genres_string = []
    genres= soup.findAll('a', class_='ipc-chip ipc-chip--on-baseAlt')
    for gn in genres:
        genres_string.append(gn.text)
    genre.append([",".join(genres_string)])
except:
    genre.append(None)
    
try:
    metascore.append(soup.find('span', class_='sc-b0901df4-0 bcQdDJ metacritic-score-box').text)
except:
    metascore.append(None)

try:
    description.append(soup.find('span', class_='sc-466bb6c-2 chnFO').text)
except:
    description.append(None)    
    
    
try:
    director_string = []
    directors= soup.findAll('li', class_='ipc-metadata-list__item')
    for di in directors[:1]:
        if di.span and di.span.text=="Director":
            directorss = di.findAll('li',class_='ipc-inline-list__item')
            for li in directorss:
                director_string.append(li.a.text)
    director.append([",".join(director_string)])
except:
    director.append(None)

try:
    writers_string = []
    writers= soup.findAll('li', class_='ipc-metadata-list__item ipc-metadata-list-item--link')
    for di in writers[:1]:
        print("di.a.text", di.a.text)
        if di.a and di.a.text in ['Writers', 'Writer']:
            writer = di.findAll('li',class_='ipc-inline-list__item')
            print("writer", writer)
            for li in writer:
                writers_string.append(li.a.text)
    writer_data.append([",".join(writers_string)])
except:
    writer_data.append(None)
print("writer_data", writer_data)
try:
    star_string = []
    stars= soup.findAll('li', class_='ipc-metadata-list__item ipc-metadata-list-item--link')
    for di in stars[:2]:
        if di.a and di.a.text=="Stars":
            star = di.findAll('li',class_='ipc-inline-list__item')

            for li in star:
                star_string.append(li.a.text)
    star_data.append(star_string)
except:
    star_data.append(None)

    
gross_data =  soup.findAll('li', class_="ipc-metadata-list__item sc-ae184607-2 jqHug")
 

try:
    bud=None
    for li in gross_data:
        if li.span and li.span.text=="Budget":
            bud=li.find('span', class_='ipc-metadata-list-item__list-content-item').text
    budget.append(brud)
except:
    budget.append(None)  

try:
    gros=None
    for li in gross_data:
        if li.span and li.span.text=="Gross worldwide":
            gros=li.find('span', class_='ipc-metadata-list-item__list-content-item').text
    gross.append(gros)
except:
    gross.append(None)    


# movie_DF =pd.DataFrame({
#     'Name of movie': movie_name,
#     'Year of relase': year,
#     'Genre': genre,
#     'Watchtime': runtime,
#     'Certificate': certificate,
#     'Movie Rating':rating,
#     'Mtascore': metascore,
#     'Director': director,
#     'Writers': writer_data,
#     'Stars': star_data,
#     'Rating': rating,
#     'Description' :description,
#     'Gross': gross,
# })
# movie_DF.to_csv("movie_DF.csv")
# print("movie_DF", movie_DF)
```

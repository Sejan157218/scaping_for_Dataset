```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

movie_name = []
genre = []
certificate=[]
year = []
runtime = []
rating = []
metascore = []
votes = []
gross = []
description = []
director = []
rating = []
writer_data = []
star_data = []

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36"}

url = 'http://www.imdb.com/title/tt9663764/?ref_=sr_t_2'

#request allow you to send HTTP request
response = requests.get(url,headers=headers)

soup = BeautifulSoup(response.content, 'html.parser')
```

```python
content_div = soup.find('div', class_="sc-e226b0e3-3 dwkouE")
movie_name.append(content_div.find('span', class_='hero__primary-text').text)
year.append(content_div.find('a', class_='ipc-link ipc-link--baseAlt ipc-link--inherit-color').text)

try:
    certificate.append(content_div.find('a', class_='ipc-link ipc-link--baseAlt ipc-link--inherit-color').text)
except:
    certificate.append(None)
    
try:
    runtime.append(content_div.find('li', class_='ipc-inline-list__item').text)
except:
    runtime.append(None)
    
try:
    rating.append(content_div.find('span', class_='sc-bde20123-1 cMEQkK').text)
except:
    rating.append(None)
    
try:
    genres_string = []
    genres= soup.findAll('a', class_='ipc-chip ipc-chip--on-baseAlt')
    for gn in genres:
        genres_string.append(gn.text)
    genre.append([",".join(genres_string)])
except:
    genre.append(None)
    
try:
    metascore.append(soup.find('span', class_='sc-b0901df4-0 bcQdDJ metacritic-score-box').text)
except:
    metascore.append(None)

try:
    description.append(soup.find('span', class_='sc-466bb6c-2 chnFO').text)
except:
    description.append(None)    
    
    
try:
    director_string = []
    directors= soup.findAll('li', class_='ipc-metadata-list__item')
    for di in directors[:1]:
        if di.span and di.span.text=="Director":
            directorss = di.findAll('li',class_='ipc-inline-list__item')
            for li in directorss:
                print("director", li.a.text)
                director_string.append(li.a.text)
    director.append([",".join(director_string)])
except:
    director.append(None)

try:
    writers_string = []
    writers= soup.findAll('li', class_='ipc-metadata-list__item ipc-metadata-list-item--link')
    for di in writers[:1]:
        if di.a and di.a.text=="Writers":
            writer = di.findAll('li',class_='ipc-inline-list__item')
            for li in writer:
                writers_string.append(li.a.text)
    writer_data.append([",".join(writers_string)])
except:
    writer_data.append(None)

try:
    star_string = []
    stars= soup.findAll('li', class_='ipc-metadata-list__item ipc-metadata-list-item--link')
    for di in stars[:2]:
        if di.a and di.a.text=="Stars":
            star = di.findAll('li',class_='ipc-inline-list__item')

            for li in star:
                star_string.append(li.a.text)
    star_data.append(star_string)
except:
    star_data.append(None)

    
gross_data =  soup.find('div', class_="sc-f65f65be-0 bBlII")
for ul in gross_data:
    print("ul", ul.li)

    
# try:
#     certificate.append(content_div_p.find('a', class_='ipc-link ipc-link--baseAlt ipc-link--inherit-color').text)
# except:
#     certificate.append(None)
# runtime.append(content_div_p.find('span', class_='ipc-inline-list__item').text if content_div_p.find('span', class_='ipc-inline-list__item') else None)
#     genre.append(content_div_p.find('span', class_='ipc-chip-list__scroller').text.replace('\n', ''))
#     rating.append(content_div.find('span', class_='ipl-rating-star__rating').text)
#     try:
#         metascore_div = content_div.find('div', class_='inline-block ratings-metascore')
#         metascore.append(metascore_div.find('span', class_='metascore favorable').text.replace('\n', ''))
#     except:
#         metascore.append(None)
#     description.append(content_div.find('p', class_='').text.replace('\n', ''))
#     de_cost = content_div.select_one('p:has(a)').text.replace('\n', '').replace('  ', '').replace('Director:','').replace('Directors:','').replace('|','').split('Stars:')
#     director.append(de_cost[0] if len(de_cost)>=1 else None)
#     stars.append(de_cost[1] if len(de_cost)>=2 else None)
#     value = store.find_all('span', attrs = {'name': 'nv'})
#     votes.append(value[0].text)
#     gross.append(value[1].text if len(value) >1 else None)

movie_DF ={
    'Name of movie': movie_name,
    'Year of relase': year,
    'Genre': genre,
    'Watchtime': runtime,
    'Certificate': certificate,
    'Movie Rating':rating,
    'Mtascore': metascore,
    'Director': director,
    'Writers': writer_data,
    'Stars': star_data,
    'Rating': rating,
    'Description' :description,
    'Votes': votes,
    'Gross': gross,
}
print("movie_DF", movie_DF)
```
